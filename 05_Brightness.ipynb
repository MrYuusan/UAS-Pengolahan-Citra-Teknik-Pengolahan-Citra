```python
# Import library skimage
import skimage
# Dari skimage import data sample
from skimage import data
# Memakai matplotlib.pyplot untuk plot gambar & disimpan dengan nama 'plt'
import matplotlib.pyplot as plt
```

```python
# Memakai gambar grass dari data skimage & disimpan dengan nama 'img'
img = data.grass()
# Menyimpan nilai row & column dari gambar grass ke num_rows & num_cols
num_rows, num_cols = img.shape
```

```python
# Tampilkan gambar asli
plt.imshow(img, cmap='gray')
```

```python
# Copy gambar
img_bright = img.copy()
# Nilai brightnes
b = 75
# Proses brightness
for i in range(num_rows):
    for j in range(num_cols):
        temp = img[i][j] + b
        if (temp > 255):
            img_bright[i][j] = 255
        elif(temp < 0):
            img_bright[i][j] = 0
        else:
            img_bright[i][j] = temp
```

```python
# Plot gambar asli dan brightness
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(img_bright, cmap='gray')
```
