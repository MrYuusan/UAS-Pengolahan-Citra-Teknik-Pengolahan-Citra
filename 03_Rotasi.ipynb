```python
# Import library math
import math
# Import library skimage
import skimage
# Dari skimage import data sample
from skimage import data
# Dari skimage import fungsi transform
from skimage import transform
# Memakai matplotlib.pyplot untuk plot gambar & disimpan dengan nama 'plt'
import matplotlib.pyplot as plt
# Memakai library numpy untuk mempermudah pembuatan array & disimpan dengan nama 'np'
import numpy as np
```

```python
# Memakai gambar camera dari data skimage & disimpan dengan nama 'img'
img = data.camera()
# Tampilkan gambar asli
plt.imshow(img, cmap='gray')
```

```python
# Membuat fungsi transformasi rotasi
rotasi = transform.SimilarityTransform(scale=1, rotation=math.pi/4,
                                      translation=(img.shape[0]/2, -100))
print(rotasi.params)
```

```python
# Proses rotasi
img_rotasi = transform.warp(img, rotasi)
# Tampilkan gambar asli & hasil rotasi
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(img_rotasi, cmap='gray')
```
