```python
# Import library skimage
import skimage
# Dari skimage import data sample
from skimage import data
# Memakai matplotlib.pyplot untuk plot gambar & disimpan dengan nama 'plt'
import matplotlib.pyplot as plt
# Memakai library numpy untuk mempermudah pembuatan array & disimpan dengan nama 'np'
import numpy as np
```

```python
# Memakai gambar coins dari data skimage & disimpan dengan nama 'img'
img = data.coins()
# Menyimpan nilai row & column dari gambar coins ke num_rows & num_cols
num_rows, num_cols = img.shape
```

```python
# Tampilkan gambar asli
plt.imshow(img, cmap='gray')
```

```python
# Proses Translasi
img_trans = np.zeros((num_rows, num_cols))
# Jarak translasi
m = 30
n = 20
for i in range(num_rows):
    for j in range(num_cols):
        if ((i+m < num_rows) and (j+n) < num_cols):
            img_trans[i][j] = img[i+m][j+n]
        else:
            img_trans[i][j] = 255
```

```python
# Tampilkan gambar asli dan hasil translasi
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(img_trans, cmap='gray')
```
