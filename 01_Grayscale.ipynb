```python
# Import library skimage
import skimage
# Dari skimage import data sample
from skimage import data
# Dari skimage import color
from skimage import color
# Memakai matplotlib.pyplot untuk plot gambar & disimpan dengan nama 'plt'
import matplotlib.pyplot as plt
```

```python
# Memakai gambar cat dari data skimage & disimpan dengan nama 'img'
img = data.cat()
```

```python
# Tampilkan gambar asli
plt.imshow(img)
```

```python
# Mengubah gambar asli menjadi grayscale & disimpan dengan nama 'img_gray'
img_gray = color.rgb2gray(data.cat())
# Tampilkan gambar asli & hasil grayscale
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(img_gray, cmap='gray')
```
